## 0. 필요한 라이브러리 불러오기

```python
import os
import glob

import pandas as pd
import numpy as np

import json
import math

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Buildings train

```python
B_train_json = os.getenv('HOME')+'/SIAIFFEL_Dataset/buildings/Training/라벨링데이터'

B_train_json = sorted(glob.glob(B_train_json + "/*json"))
len(B_train_json)
```

__EDA를 위한 csv변환__

1. Site.csv
2. Road.csv
3. Building.csv
4. BnR.csv

```python
B_train_json
```

```python
B_train_list = []

for i in B_train_json:
    B_json_feature = json.load(open(i))
    
    for j in B_json_feature['features']:
        
        # Geometry 
        geo_coord = j['geometry']['coordinates']
        geo_type = j['geometry']['type']
        
        # Properties
        type_id = j['properties']['type_id']
        type_name = j['properties']['type_name']
        object_imcoords = j['properties']['object_imcoords']
        building_imcoords = j['properties']['building_imcoords']
        road_imcoords = j['properties']['road_imcoords']
        image_id = j['properties']['image_id']
        ingest_time = j['properties']['ingest_time']
        
        # append B_train_list
        B_train_list.append([image_id, type_name, type_id, geo_coord, geo_type,
                           object_imcoords, building_imcoords, road_imcoords, ingest_time])
```

```python
# json to csv
B_train = pd.DataFrame(B_train_list,
                        columns = ['image_id','type_name','type_id','geo_coord','geo_type',
                                   'object_imcoords','building_imcoords','road_imcoords','ingest_time'])
```

```python
B_train
```

```python
# NA check
nan = B_train[(B_train['building_imcoords'] == '')].index
nan
```

```python
# building_imcoords dataframe
B_train_coord = pd.DataFrame(columns=['building_imcoords'])
B_train_coord
```

```python
B_train['building_imcoords'][149300]
```

```python
len(B_train.index)
```

```python
# round building_imcoords dataframe

for i in range(len(B_train.index)): # 149300
    
    polygon = B_train['building_imcoords'][i]
    
    if polygon == '':
        
        B_train_coord = B_train_coord.append({'building_imcoords': polygon}, ignore_index=True)
    
    else:
        
        polygon = list(map(float, polygon.split(',')))
        polygon = list(map(round, polygon))
        polygon = np.array(polygon, np.int32).reshape(-1)
    
        B_train_coord = B_train_coord.append({'building_imcoords': polygon}, ignore_index=True)
#     print(df)
    
B_train_coord
```

```python
# image_id dataframe

B_train_image_id = pd.DataFrame(B_train['image_id'])
B_train_image_id
```

```python
# concat B_train_image_id, building_imcoords

Buildings_train = pd.concat([B_train_image_id, B_train_coord], axis=1)
Buildings_train = pd.DataFrame(Buildings_train)
Buildings_train
```

```python
# ignore NA
Buildings_train.iloc[17441]
```
