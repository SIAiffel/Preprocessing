### import tiff with cv2

```python
import cv2

img = cv2.imread(spacenet_sample)
print(img.min(), img.max())
plt.imshow(img)
```

```python
plt.imshow(img)
```

```python
img_scaled = cv2.normalize(img, dst=None, alpha=55555, beta=0, norm_type=cv2.NORM_MINMAX)
plt.imshow(img_scaled)
```

```python
img_scaled = cv2.normalize(img, dst=None, alpha=30, beta=40, norm_type=cv2.NORM_MINMAX)
plt.imshow(img_scaled)
```
