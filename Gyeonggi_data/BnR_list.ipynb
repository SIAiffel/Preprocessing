```python
import math
import os
import glob
import json
```

```python
train_BnR = os.getenv('HOME')+'/Seoul_dataset/train/label/BnR'

train_BnR_png = sorted(glob.glob(train_BnR + "/*png"))
len(train_BnR_png)
```

```python
train_BnR_png
```

```python
train_BnR_list = []

for i in train_BnR_png:
    train_BnR_list.append(i.split('/')[-1])
train_BnR_list
```

```python
train_img = os.getenv('HOME')+'/Seoul_dataset/train/image'

train_png = sorted(glob.glob(train_img + "/*png"))
len(train_png)
```

```python
import os
import shutil

oPath = os.getenv('HOME')+'/Seoul_dataset/train/image/'
nPath = os.getenv('HOME')+'/Seoul_dataset/train/non_BnR/'

# 이동할 파일들을 닮을 배열 객체
mov_files = []

# Files List Create - TEXT 파일을 제외한 파일들이 이동대상 파일
for file in train_BnR_list:
    if nPath.split('/')[-1] == file:
        print("BnR File : " + file)
    else:
        print("Else File : " + file)
        mov_files.append(file)

# Move Files
for movfi in mov_files:
    print('Move File : ' + oPath + ' 에 있는 ' + movfi + ' ----- > ' + nPath + '로 이동하였습니다.')
    shutil.move(oPath + '/' + movfi, nPath + '/' + movfi)
```
