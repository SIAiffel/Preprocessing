## Roads train

```python
import os
import glob

import pandas as pd
import numpy as np

import json
import math

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
R_train_json = os.getenv('HOME')+'/aiffel/siaiffel/satellite_data/roads_training/라벨링데이터'

R_train_json = sorted(glob.glob(R_train_json + "/*json"))
len(R_train_json)
```

```python
R_train_json
```

```python
R_train_list = []

for i in R_train_json:
    R_json_feature = json.load(open(i))
    
    for j in R_json_feature['features']:
        
        # Geometry 
        geo_coord = j['geometry']['coordinates']
        geo_type = j['geometry']['type']
        
        # Properties
        type_id = j['properties']['type_id']
        type_name = j['properties']['type_name']
        object_imcoords = j['properties']['object_imcoords']
        building_imcoords = j['properties']['building_imcoords']
        road_imcoords = j['properties']['road_imcoords']
        image_id = j['properties']['image_id']
        ingest_time = j['properties']['ingest_time']
        
        # append B_train_list
        R_train_list.append([image_id, type_name, type_id, geo_coord, geo_type,
                           object_imcoords, building_imcoords, road_imcoords, ingest_time])
```

```python
R_train = pd.DataFrame(R_train_list,
                        columns = ['image_id','type_name','type_id','geo_coord','geo_type',
                                   'object_imcoords','building_imcoords','road_imcoords','ingest_time'])
```

```python
R_train
```

```python
# NA check
nan = R_train[(R_train['road_imcoords'] == '')].index
nan
```

```python
# building_imcoords dataframe
R_train_coord = pd.DataFrame(columns=['road_imcoords'])
R_train_coord
```

```python
R_train['road_imcoords'][23386]
```

```python
len(R_train.index)
```

```python
# round building_imcoords dataframe

for i in range(len(R_train.index)): # 149300
    
    polygon = R_train['road_imcoords'][i]
    
    if polygon == '':
        
        R_train_coord = R_train_coord.append({'road_imcoords': polygon}, ignore_index=True)
    
    else:
        
        polygon = list(map(float, polygon.split(',')))
        polygon = list(map(round, polygon))
        polygon = np.array(polygon, np.int32).reshape(-1)
    
        R_train_coord = R_train_coord.append({'road_imcoords': polygon}, ignore_index=True)
#     print(df)
    
R_train_coord
```

```python
# image_id dataframe

R_train_image_id = pd.DataFrame(R_train['image_id'])
R_train_image_id
```

```python
# concat B_train_image_id, building_imcoords

Roads_train = pd.concat([R_train_image_id, R_train_coord], axis=1)
Roads_train = pd.DataFrame(Roads_train)
Roads_train
```
